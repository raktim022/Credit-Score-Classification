# Data Cleaning and Preprocessing

```python
#Dropping columns that do not seem relevant or contribute to the credit score
df=df.drop(columns=['Month','Name','SSN','ID','Customer_ID','Type_of_Loan'])
```

<b>Dropping null rows</b>

```python
df=df.dropna(subset='Credit_History_Age')
```

```python
nil=df['Occupation'].unique()[1]
df=df[df['Occupation']!=nil]
df.reset_index(drop=True,inplace=True)
```

```python
#Converting Credit History to a numerical attribute
import re
for i in range(len(df)):
    a,b=re.findall(r'\d+',str(df.loc[i,'Credit_History_Age']))
    df.loc[i,'Credit_History_Age']=int(a)+(int(b)/12)
```

```python
df.reset_index(inplace=True)
```

```python
df['Monthly_Inhand_Salary'].mean()
```

```python
import numpy as np
for i in range(len(df)):
    if str(df.loc[i,'Monthly_Inhand_Salary'])=='nan':
        df.loc[i,'Monthly_Inhand_Salary']=df['Monthly_Inhand_Salary'].mean()
```

```python
#Dropping null values
df=df.dropna(subset=['Num_Credit_Inquiries'])
df.reset_index(drop=True,inplace=True)
```

```python
df=df.dropna(subset=['Monthly_Balance'])
df.reset_index(drop=True,inplace=True)
```

<b>Removing Amount_invested_monthly and Num_of_Delayed_Payment because most of it's values are 0</b>

```python
for i in range(len(df)):
        val=str(df.loc[i,'Amount_invested_monthly'])
        try:
            df.loc[i,'Amount_invested_monthly']=float(re.findall(r'\d+',val)[0])
        except:
            df.loc[i,'Amount_invested_monthly']=0
```

```python
df['Amount_invested_monthly'].hist()
```

```python
df=df.drop(columns=['Amount_invested_monthly'])
```

```python
#Dropping columns that will not be needed in building the model
df=df.drop(columns=['level_0','index'])
```

```python
df['Num_of_Delayed_Payment'].isnull().sum()
```

```python
for i in range(len(df)):
        val=str(df.loc[i,'Num_of_Delayed_Payment'])
        try:
            df.loc[i,'Num_of_Delayed_Payment']=float(re.findall(r'\d+',val)[0])
        except:
            df.loc[i,'Num_of_Delayed_Payment']=0
```

```python
df['Num_of_Delayed_Payment'].hist()
```

```python
df=df.drop(columns='Num_of_Delayed_Payment')
df.reset_index(inplace=True,drop=True)
```

```python
df.isnull().sum()
```

# Handling Data Imbalance

```python
df['Credit_Score'].value_counts()
```

<b>Oversampling the 'Poor' and 'Good' labels to have the same number of training values as 'Standard'</b>

```python
from sklearn.utils import resample
good_credit=df[df['Credit_Score']=='Good']
standard_credit=df[df['Credit_Score']=='Standard']
poor_credit=df[df['Credit_Score']=='Poor']
poor_credit_resamp=resample(poor_credit,replace=True,n_samples=4666)
good_credit_resamp=resample(good_credit,replace=True,n_samples=4666)
df_merged=pd.concat([poor_credit_resamp,standard_credit,good_credit_resamp])
df_merged
```

# Scaling and converting categorical attributes to numerical attributes

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col_enc=['Occupation','Credit_Mix','Payment_of_Min_Amount','Payment_Behaviour','Credit_Score']
for col in col_enc:
    df_merged[col]=le.fit_transform(df_merged[col])
df_merged.reset_index(inplace=True,drop=True)
```

```python
df_merged['Credit_Score'].value_counts()
```

```python
#Columns with numerical attributes stored as strings
cols_str=['Age','Annual_Income','Outstanding_Debt','Monthly_Balance','Num_of_Loan']
```

```python
#Converting strings to numerical values
for i in range(len(df_merged)):
    for col in cols_str:
        a=float(re.findall(r'\d+',str(df_merged.loc[i,col]))[0])
        df_merged.loc[i,col]=a
```

```python
df_merged
```

<b>Scaling the data</b>

```python
col_scl=['Age','Annual_Income','Monthly_Inhand_Salary','Num_Bank_Accounts','Num_of_Loan',
        'Num_Credit_Card','Interest_Rate','Delay_from_due_date','Changed_Credit_Limit',
        'Num_Credit_Inquiries','Outstanding_Debt','Credit_Utilization_Ratio','Credit_History_Age',
        'Total_EMI_per_month','Monthly_Balance']
```

```python
for col in col_scl:
    try:
        df_merged[col]=(df_merged[col]-df_merged[col].mean())/df_merged[col].std()
    except TypeError:
        print(col,'has corrupted numerical data')
```

```python
df_merged=df_merged.drop(columns=['Changed_Credit_Limit'])
df_merged.reset_index(drop=True,inplace=True)
```

```python
df_merged.head()
```

```python
#Splitting the datast into input and labels
y=df_merged['Credit_Score']
X=df_merged.drop(columns=['Credit_Score'])
```

# Building the model

<b>Using Random Forest classifier to predict the credit score</b>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape,X.shape,y.shape
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
pred=rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import plot_confusion_matrix
confusion_matrix = plot_confusion_matrix(rf, X_test, y_test)
```

<b>Using Decision Tree classifier to predict the credit score</b>

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)

pred=clf.predict(X_test)


print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import plot_confusion_matrix
confusion_matrix = plot_confusion_matrix(clf, X_test, y_test)
```
